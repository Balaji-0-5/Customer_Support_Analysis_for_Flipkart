## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# dropping connected_handling_time
data.drop('connected_handling_time',axis=1,inplace=True)
````

````python
# ml-model for sentiment analysis of Customer Remarks
# https://huggingface.co/cardiffnlp/twitter-xlm-roberta-base-sentiment
from transformers import pipeline
model_path = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
sentiment_task = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path)
````

````python
#Example
sentiment_task("not nice very pathetic")[0]['label']
````

````python
#Create a new Sentiment column with the results of sentiment analysis
# Warning : this line runs ml model (use gpu) also it might take about 10 to 15 mins
data["Sentiment"] = data["Customer Remarks"].astype(str).apply(lambda x : sentiment_task(x[:400])[0]['label'])
````

````python
#for manual checking
data[["Customer Remarks","Sentiment"]].dropna().head(20)
````

````python
data.to_csv('Customer_support_data_cleaned.csv',index=False)
````

````python
data_cleaned = pd.read_csv('Customer_support_data_cleaned.csv')
````

````python
date_cols = ['order_date_time',
 'Issue_reported at',
 'issue_responded',
 'Survey_response_Date']
````

````python
data_cleaned.fillna({"Product_category":"Unspecified"},inplace=True)
````

````python
for col in date_cols:
  data_cleaned[col] = pd.to_datetime(data_cleaned[col] ,dayfirst=False ,errors='coerce')
````

````python
data_cleaned["Response Time (mins)"] = (data_cleaned["issue_responded"] - data_cleaned["Issue_reported at"]).dt.total_seconds() / 60
````

````python
data_cleaned.info()
````

### What all manipulations have you done and insights you found?

   - Extracted the sentiment label (Positive, Neutral, or Negative) and stored it in a new column **`Sentiment`**.

3. **Handled Missing Values:**
   - Replaced missing values in **`Product_category`** with `"Unspecified"` to avoid loss of data.

     - `Survey_response_Date`

````python
# Chart - 7 visualization code
# Compute proportions
csat_proportions = pd.crosstab(data_cleaned["Tenure Bucket"], data_cleaned["CSAT Score"], normalize="index")

# Convert to long format for Seaborn
csat_proportions = csat_proportions.reset_index().melt(id_vars="Tenure Bucket", var_name="CSAT Score", value_name="Proportion")

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=csat_proportions, x="Tenure Bucket", y="Proportion", hue="CSAT Score", palette="coolwarm_r")

plt.title("Proportional CSAT Score Distribution by Tenure Bucket",y=1.15,fontsize=15)
plt.ylabel("Proportion",fontsize=15)
plt.xlabel("Tenure Bucket",fontsize=15)
plt.legend(title="CSAT Score",bbox_to_anchor=(0.2,1),ncol=5)
plt.show()
````

##### 1. Why did you pick the specific chart?

This **grouped bar chart** effectively showcases how CSAT scores are distributed across different tenure buckets. By using **proportions**, it ensures that differences in sample size don't distort the comparison between tenure groups.

#### Chart - 11

````python
# Chart - 11 visualization code
# Complaints by Category
category_counts = data_cleaned["category"].value_counts().sort_values(ascending=False)
category_cumulative = np.cumsum(category_counts) / category_counts.sum() * 100

fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart
sns.barplot(x=category_counts.index, y=category_counts.values, ax=ax1)
ax1.set_ylabel("Complaint Count")
ax1.set_xlabel("Category")
ax1.set_title("Pareto Analysis: Complaints by Category",y=1.05,fontsize=15)
ax1.tick_params(axis="x", rotation=45)

# Cumulative percentage line
ax2 = ax1.twinx()
ax2.plot(category_counts.index, category_cumulative, color="r", marker="o", linestyle="dashed", label="Cumulative %")
ax2.axhline(y=80, color="gray", linestyle="dotted")  # 80% threshold line
ax2.set_ylabel("Cumulative Percentage")
ax2.legend(bbox_to_anchor=(0.6, 1.07))

plt.show()
````

##### 1. Why did you pick the specific chart?

A **Pareto chart** is used to identify the most significant factors contributing to an issue. It follows the **80/20 rule**, showing that a small number of categories account for most complaints. It is used here to understand the major categories causing customer complaints.


##### 2. What is/are the insight(s) found from the chart?

- **Returns and Order-related complaints** contribute to the majority of total complaints.
- The cumulative percentage curve (red) shows that addressing the top two categories would resolve nearly **80% of complaints**.
- Other categories, like **refund-related, product queries, and Shopzilla-related issues**, have a smaller but still notable impact.
